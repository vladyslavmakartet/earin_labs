```python
import gym 
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import VecFrameStack, DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
import os
# from sklearn.model_selection import ParameterGrid # To add
```

# 2. Test Environment

```python
environment_name = "CarRacing-v0"
```

```python
env = gym.make(environment_name)
```

```python
episodes = 5
for episode in range(1, episodes+1):
    state = env.reset()
    done = False
    score = 0 
    
    while not done:
        env.render()
        action = env.action_space.sample()
        n_state, reward, done, info = env.step(action)
        score+=reward
    print('Episode:{} Score:{}'.format(episode, score))
env.close()
```

```python
env.close()
```

# 3. Train Model

```python
log_path = os.path.join('Training', 'Logs')
```

```python
model = PPO("CnnPolicy", env, verbose=1, tensorboard_log=log_path)
```

```python
model.learn(total_timesteps=256000)
```

# 4. Save Model 

```python
ppo_path = os.path.join('Training', 'Saved Models', 'PPO_Driving_model_256k')
```

```python
model.save(ppo_path)
```

# 5. Evaluate and Test

```python
print(evaluate_policy(model, env, n_eval_episodes=20, render=True))
```

```python
env.close()
```

```python
obs = env.reset()
while True:
    action, _states = model.predict(obs.copy())
    obs, rewards, dones, info = env.step(action)
    env.render()
```
